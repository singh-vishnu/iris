```python
x=data.drop('Species',axis=1)
y=data['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# logistice regression
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(ytest,ypred)
```

```python
lr_df=pd.DataFrame()
lr_df['y_actual']=ytest
lr_df['y_model']=ypred
lr_df
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python

dt.fit(xtrain,ytrain)
```

```python

ypred_dt=dt.predict(xtest)
```

```python
accuracy_score(ytest,ypred_dt)
```

```python
# random forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
ypred_rfc=rfc.predict(xtest)
```

```python
accuracy_score(ytest,ypred_rfc)
```

```python
rfc_df=pd.DataFrame()
rfc_df['yactual']=ytest
rfc_df['ymodeled']=ypred_rfc
rfc_df
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(xtrain,ytrain)
svc_pred=svc.predict(xtest)
```

```python
accuracy_score(ytest,svc_pred)
```
